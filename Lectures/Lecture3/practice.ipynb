# Activities

Note: the practice happens almost exclusively on paper, potentially augmented by multiple choice questions in order to submit an answer that shows understanding.

Expression evaluation for numerous boolean expressions with state:
    - `eval_expr(<x>,{x:=False})`
        - possible answers:
            - x
            - {x:= False}[x]
            - False
            - True
    - `eval_expr(<x and y>,{x:=True, y:=False})`
        - possible answers:
            - {x:=False, y:=True}[x] and {x:=False, y:=True}[y]
            - `<x and False>`
            - False
            - `<True and y>`
    - `eval_expr(<x != y>,{x:=True, y:=False})`
        - possible answers:
            - `<True != y>`
            - `<x != False>`
            - False
            - True
    - `eval_expr(<a and b and c and d>,{d:=False, c:=True, b:=True, a:=True})`
        - possible answers:
            - `<a and b and c and False>`
            - `<True and c and d>`
            - False
            - `<True and b and c and d>`
    - `eval_expr(<a or b or c or d>,{d:=False, c:=True, b:=False, a:=False})`
        - possible answers:
            - `<a or b or c or False>`
            - `<False or c or d>`
            - `<False or b or c or d>`
            - True            
    - `eval_expr(<((a or u) and not (b and not v)) or ((c and not w) or (d and not x))>,{d:=False, c:=True, b:=True, a:=True, u:=False, v:=False, w:=False, x:=True})`
        - possible answers:
            - False
            - True
            - `<((True or u) and not (b and not v)) or ((c and not w) or (d and not x))>`
            - `<((True or False) and not (True and not False)) or ((True and not False) or (False and not True))>`

    - `eval_expr(<(10 // 3)>,{})`

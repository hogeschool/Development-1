- we now move on to some actual Python coding
- we will explore some of the minimal differences in syntax
    - focus on learning how to abstract the syntax away: it may seem important now, but it is only decoration and as such fulfills no substantial role
    - we do this on purpose: learn to think in terms of a (clean) abstract model, the language itself is only a "minor detail" as we need something to code into 
- assignment in Python becomes `V = E`, where `V` is a variable name and `E` is an expression
    - give a few examples of assignments
    - so far we used `:=`, notice that with the Python syntax we can write mathematical nonsense like `x = x + 1`
- expressions are essentially the same as what we have seen so far on paper: 
    - `A + B`
    - `A - B`
    - `A * B`
    - `A // B` (integer)
    - `A / B` (float)
    - `A % B`
    - `-A` (negation)
    - `A and B`
    - `A or B`
    - `not A`
- operator precedence is the same as we have seen so far:
    - multiplication and division
    - sum and subtraction
    - [optional]: and has higher precedence than or, why is that logical wrt arithmetic operator precedence?

- let us see some examples of operators

```
print(True and True)
print(True and False)
print(not True and False)
print(not (True and False))
print(True or True)
print(True or False)
print(3 + 2 * 5)
print(5 // 2)
print(5 / 2)
```
    
- input is read with `input()`
    - `int(input())` reads input
    - if between the `input` brackets we set a string, then it is shown to the user
- output is printed with `print()`
- `str(E)` converts an expression `E` to a string
- conditionals use the colon as a delimiter; where we wrote 

```
if c then
  P
else 
  q
```

- we now write 

```
if c:
  p
else:
  q
```

- for example (discuss the problem of building a tiny calculator, and build this interactively in class):

```
n = int(input("insert the first operand\n"))
op = input("insert the operation (+,-,*,/)\n")
m = int(input("insert the second operand\n"))
if op == "+":
  res = n + m
elif op == "-":
  res = n - m
elif op == "*":
  res = n * m
else:
  res = n // m
print(f"{n}{op}{m} -> {res}")
```

- indentation stil defines where a sub-program begins and ends

- loops are the same, again with colons as delimiters:

```
while c do
  b
```

- becomes, in Python

```
while c:
  b
```

- for example (discuss the integer logarithm, and build it interactively in class)

```
n = int(input("insert the number to divide\n"))
b = int(input("insert the divisor\n"))
i = 0
m = n
while m >= b:
  m = m // b
  i = i + 1
print(f"log_{b}({n}) = {i}")
```

- for example (discuss the integer elevation to power, and build it interactively in class)

```
b = int(input("insert the base\n"))
n = int(input("insert the exponent\n"))
i = 0
m = 1
while i < n:
  m = m * b
  i = i + 1
print(f"{b}^{n} = {m}")
```
